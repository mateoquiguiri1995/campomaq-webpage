```python
import os
from dotenv import load_dotenv, find_dotenv
import pandas as pd
from pymongo import MongoClient
from pymongo.server_api import ServerApi
# Load environment variables
load_dotenv(find_dotenv())
```

```python
# MongoDB connection
client = MongoClient(os.environ['MONGO_URI'], server_api=ServerApi('1'))
db = client["campomaq"]

# Create or get synonyms collection
synonyms_collection = db["cm_synonyms"]

# Synonym mappings
synonym_docs = [
    {
        "mappingType": "equivalent",   # interchangeable
        "synonyms": ["motoguadaña", "desbrozadora"]
    },
    {
        "mappingType": "equivalent",
        "synonyms": ["corredor de altura", "cortasetos"]
    },
    {
        "mappingType": "explicit",
        "input": ["bomba de fumigar"],
        "synonyms": ["fumigadora", "bomba de fumigacion", "fumigadora motorizada"]
    }
]

# Insert all at once
result = synonyms_collection.insert_many(synonym_docs)

print(f"Inserted synonym mappings with _ids: {result.inserted_ids}")
```
