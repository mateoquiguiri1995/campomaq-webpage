```python
import openai
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
from tqdm import tqdm
import math
import json
from concurrent.futures import ThreadPoolExecutor, as_completed


# Load environment variables
load_dotenv(find_dotenv())
tqdm.pandas()

openai.api_key = os.environ['OPENAI_API_KEY']
# Initialize client
client = OpenAI()


# Function to process one batch
def process_batch(i, batch, num_batches):
    prompt = "Write concise and appealing product descriptions for an agricultural machinery e-commerce site.\n\n"
    for idx, row in batch.iterrows():
        prompt += f"Product {idx}:\n"
        prompt += f"Name: {row['product_name']}\n"
        prompt += f"Category: {row['category_name']}\n"
        prompt += f"Brand: {row['brand_name']}\n\n"
    prompt += "Return the results as JSON with keys being the product index and values being the description."

    response = client.chat.completions.create(
        model="gpt-5-nano",  # you can switch model here
        messages=[{"role": "user", "content": prompt}],
        # temperature=0.7
    )

    try:
        batch_desc = json.loads(response.choices[0].message.content)
    except Exception as e:
        print(f"⚠️ JSON parsing failed in batch {i}: {e}")
        batch_desc = {str(idx): response.choices[0].message.content for idx in batch.index}

    # 👀 Print one sample from this batch
    sample_idx = next(iter(batch.index))
    print(f"\nBatch {i+1}/{num_batches} sample → {batch_desc.get(str(sample_idx), '')[:150]}...\n")

    return batch_desc

# Main function
def generate_descriptions_multithread(df, batch_size=20, max_workers=5):
    descriptions = {}
    num_batches = math.ceil(len(df) / batch_size)

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        for i in range(num_batches):
            batch = df.iloc[i*batch_size:(i+1)*batch_size]
            futures.append(executor.submit(process_batch, i, batch, num_batches))

        for future in tqdm(as_completed(futures), total=len(futures), desc="Generating descriptions"):
            batch_desc = future.result()
            for k, v in batch_desc.items():
                descriptions[int(k)] = v

    return [descriptions[idx] for idx in df.index]

# Run with multithreading
final_df["description"] = generate_descriptions_multithread(final_df, batch_size=20, max_workers=5)
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import math

def process_embedding_batch(i, texts, num_batches):
    response = client.embeddings.create(
        model="text-embedding-3-large",   # or "text-embedding-3-large" for higher accuracy
        input=texts,
        dimensions=256
    )
    embeddings = [item.embedding for item in response.data]

    # 👀 Print one example vector length per batch
    print(f"Batch {i+1}/{num_batches} → got {len(embeddings)} embeddings, dim={len(embeddings[0])}")

    return embeddings

def generate_embeddings_multithread(df, batch_size=50, max_workers=5):
    """
    Generate embeddings for product_name + category_name + brand_name
    using multithreaded batching.
    """
    texts = [
        f"{row['product_name']} | {row['category_name']} | {row['brand_name']}"
        for _, row in df.iterrows()
    ]

    embeddings = [None] * len(texts)
    num_batches = math.ceil(len(texts) / batch_size)

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        for i in range(num_batches):
            batch_texts = texts[i*batch_size:(i+1)*batch_size]
            futures.append(executor.submit(process_embedding_batch, i, batch_texts, num_batches))

        batch_start = 0
        for future in tqdm(as_completed(futures), total=len(futures), desc="Generating embeddings"):
            batch_embeddings = future.result()
            # Insert results back into correct positions
            embeddings[batch_start:batch_start+len(batch_embeddings)] = batch_embeddings
            batch_start += len(batch_embeddings)

    return embeddings

# Run and add embeddings column
final_df["embedding"] = generate_embeddings_multithread(final_df, batch_size=30, max_workers=5)
```

# New field to indicate that the product is shown in the catalogm

```python
# Select the database and collection
cm_env = os.environ.get('CM_ENV', 'dev')
if cm_env == 'prod':
    collection = client["campomaq"]["cm_catalog"]
else:
    collection = client["campomaq_test"]["cm_catalog"]

# New field to indicate that the product is shown in the catalog
top_100_codes = (
    df_top_sales_prods.sort_values("popularity", ascending=False)
    .head(100)["product_code"]
    .tolist()
)

# Assign True for top 100, False for others
df_top_sales_prods["show_in_app"] = df_top_sales_prods["product_code"].isin(top_100_codes)

from pymongo import UpdateOne

# Build bulk operations
operations = [
    UpdateOne(
        {"product_code": row["product_code"]},
        {"$set": {"show_in_app": bool(row["show_in_app"])}}
    )
    for _, row in df_top_sales_prods.iterrows()
]
# Execute in bulk
if operations:
    result = collection.bulk_write(operations, ordered=False)
    print(f"Matched: {result.matched_count}, Modified: {result.modified_count}")

```
